# BSP BEP-Berechnung
Die BEP Berechnung befasst sich mit dem Thema ab wann(gemessen an Anzahl verkaufter Produkte (x) oder gemessen an generiertem Umsatz (S)) ein Produkt kostendeckend ist. D.h. der BEP ist der Punkt, an welchem ich alle bisher ausgegebenen Kosten aus FC&VC wieder erwirtschaftet habe

```python
# Berechnung BEP gemessen an Verkauften Produkten
# Reihenfolge(TFC, VC, SP)
BEPu = CostFunctionContainer.calBEPUnits(TFC,VC,SP)
print(BEPu)
```

```python
# Berechnung BEP-Sales
# BEP Saleszeigt auf wieviel Umsatz erzeugt werden muss
# bevor ich den ersten Dollar erschaffe

# 1.te Möglichkeit = BEPu * SP
BEPs1 = (BEPu * 400)
print(BEPs1)

# 2.te Möglichkeit = TFC / CMR
BEPs2 = CostFunctionContainer.calBEPSalesWithCMR(TFC, VC, SP)
print(BEPs2)
```

Sobald <code>BEPu</code> Einheiten Verkauft wurden, ist mit einem Gewinn zu rechnen, hier sind die aktuellen VC & TFC = 0

```python
# Check-up Breakeven via Calc-Profit, 
# Am break even ist der Profit = 0
# Reihenfolge (X, TFC, VC, SP, switch) // switch 1/0 = mit/ohneIntegral
CostFunctionContainer.calcProfit(BEPu,TFC,VC,SP)
```

```python
# Plotten der BEP-Analytics
CostFunctionContainer.printBEP(X, TFC, VC, SP, 0)
```
